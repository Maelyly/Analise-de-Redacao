```
import pandas as pd

data = pd.read_csv("/content/Banco_redacoes_base - Banco_redacoes.csv")
data
```

```
texto, classe = data["Redacao"] , data["Nota"]
```

## Remoção Stop Words

```
import spacy
import os
os.system("python -m spacy download pt")
```

```
nlp = spacy.load('pt')
```

```
FraseSemStopWord= []
for frase in texto:
  doc = nlp(frase)
  new = [token.text.lower() for token in doc if not(token.is_stop or token.is_punct)] 
  new = " ".join(new)
  
  ##Lemmatization
  lemmas = [token.lemma_.lower() for token in nlp(new)]
  lemmas = " ".join(lemmas)
  FraseSemStopWord.append(lemmas)
```

## Array com Frases sem Stop Word

```
#FraseSemStopWord
```

##  Texto e Classes Teste e Treino

```
Classes = []
for frase in classe:
  new = frase
  if new is not 'Nota':
    Classes.append(new)

#map(int, Classes) 
```

```
Classes
```

Divisão de treinamento e teste

```
treino_texto = FraseSemStopWord[1:1900]
teste_texto = FraseSemStopWord[1901:2159]
treino_classe = Classes[1:1900]
teste_classe = Classes[1901:2159]
teste_redacao =  FraseSemStopWord[2160:2164]
teste_nota = Classes[2160:2164]
```

# Extração de Caracteristicas

###Frequência

```
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(encoding='latin-1')
X_treino_counts = count_vect.fit_transform(treino_texto) # OBS: concatenar array de stop words!!!
X_treino_counts.shape
```

###TD-IDF

```
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer(use_idf=True)#decide se vai usar o TF ou TF-IDF
X_train_tfidf = tfidf_transformer.fit_transform(X_treino_counts)
X_train_tfidf.shape
```

#Classificação

###Naive Bayes

```
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, treino_classe)
```

Avaliação - treino/teste

```
from sklearn.metrics import accuracy_score
X_teste_counts = count_vect.transform(teste_texto)
X_teste_tfidf = tfidf_transformer.transform(X_teste_counts)
predito = clf.predict(X_teste_tfidf)
accuracy_score(teste_classe, predito)
```

```
from sklearn import metrics
print(metrics.classification_report(teste_classe, predito))
```

Matriz de Confusão

```
metrics.confusion_matrix(teste_classe, predito)
```
